Let's talk more about an essential property of the *dot product*. As you know, with the dot product we take each corresponding coordinate, multiply, and then sum the products. That is: $u \cdot v = \sum_i u_i v_i$.

Let's talk about what this means *geometrically*. Consider a vector $u = (u_1, u_2)$. Let's talk about how to *decompose* $u$ onto $e_1 = (1,0), e_2 = (0, 1)$. $e_1, e_2$ are called *coordinate vectors*. Decompose means "write as a weighted sum." In the case of projecting $u$ onto $e_1, e_2$, the answer is simple: $u = u_1 e_1 + u_2 e_2$.

It is also clear that the projection of $u + v$ onto $e_1, e_2$ is equal to $(u_1 + v_1)e_1 + (u_2 + v_2)e_2$. And also: $\alpha u$ decomposes to $\alpha u_1 e_1 + \alpha u_2 e_2$. The combination of these two properties makes decomposition a *linear operation*.

Decomposition into coordinate vectors is easy (provided $u$ is specified as a coordinate pair). But we often want to decompose a vector $u$ into a pair of non-coordinate vectors $v, w$. In particular, we often want to write $u$ as a sum of *perpindicular* vectors $v, w$. This will tell us how much of $u$ is *parallel* to $v$ (perpindicular to $w$) and how much is *perpindicular* to $v$ (parallel to $w$).

What I want to note is that projection is *symmetric*.

When we only care about how much of $u$ is parallel to $v$ (and don't care about $w$), this amount is called the *projection* of $u$ onto $v$.
